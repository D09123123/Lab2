```python
import pygame
import time

WIN_WIDTH = 1024
WIN_HEIGHT = 600
BTN_WIDTH = 80
BTN_HEIGHT = 80
HP_WIDTH = 40
HP_HEIGHT = 40
FPS = 30

# color (RGB)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# initialization
pygame.init()
# load image (background, enemy, buttons, word type, start_ticks)
background_image = pygame.transform.scale(pygame.image.load("images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))
continue_image = pygame.transform.scale(pygame.image.load("images/continue.png"), (BTN_WIDTH, BTN_HEIGHT))
enemy_image = pygame.transform.scale(pygame.image.load("images/enemy.png"), (60, 60))
hp_image = pygame.transform.scale(pygame.image.load("images/hp.png"), (HP_WIDTH, HP_HEIGHT))
hp_gray_image = pygame.transform.scale(pygame.image.load("images/hp_gray.png"), (HP_WIDTH, HP_HEIGHT))
muse_image = pygame.transform.scale(pygame.image.load("images/muse.png"), (BTN_WIDTH, BTN_HEIGHT))
pause_image = pygame.transform.scale(pygame.image.load("images/pause.png"), (BTN_WIDTH, BTN_HEIGHT))
sound_image = pygame.transform.scale(pygame.image.load("images/sound.png"), (BTN_WIDTH, BTN_HEIGHT))
font = pygame.font.SysFont("arial", 20)
start_ticks = 0
# ...(to be done)


# set the title (file name)
pygame.display.set_caption("My first game")


class Game:
    def __init__(self):
        # window
        self.window = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))
        self.clock = pygame.time.Clock()                               #載入畫面 時間函數
        
        # ...(to be done)

        # hp
        self.hp = 7
        self.max_hp = 10
        pass

    def game_run(self):
        # game loop
        run = True
        while run:
            # event loop
            self.clock.tick(FPS)
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    run = False
            # ... (to be done)

            # draw background
            self.window.blit(background_image, (0, 0))  #放背景
            pygame.draw.rect(self.window, BLACK, [0, 0 , WIN_WIDTH, 95])  #放資訊欄的背景
            title = font.render("My first game", True, WHITE, BLACK)  #寫 My first game
            self.window.blit(title, (435, 0))  #放 My first game字體在資訊欄上
            # draw enemy and health bar
            self.window.blit(enemy_image, (50, 250))  #放敵人
            pygame.draw.rect(self.window, RED, [50, 250 , 60, 5])  #畫血條
#             # ...(to be done)

            # draw menu (and buttons)
             
            #400~550 width 30,50 height
            for full_hp in range(self.max_hp):  #畫總血量
                if full_hp < 5:
                    self.window.blit(hp_gray_image, (400+30*full_hp, 30))  #第一行5hp 超過 5 hp 要換行
                else:
                    self.window.blit(hp_gray_image, (400+30*(full_hp-5), 60))   #第二行的 5 hp (6~10)
            for currrent_hp in range(self.hp):  #畫現有血量
                if currrent_hp < 5:
                    self.window.blit(hp_image, (400+30*currrent_hp, 30))   #第一行5hp 超過 5 hp 要換行
                else:
                    self.window.blit(hp_image, (400+30*(currrent_hp-5), 60))    ##第二行的 5 hp (6~10)
                
            self.window.blit(muse_image, (625, 10))                        #放靜音鍵
            self.window.blit(sound_image, (725, 10))                       #放開聲音鍵
            self.window.blit(continue_image, (825, 10))                    #放繼續鍵
            self.window.blit(pause_image, (925, 10))                       #放暫停鍵
            
            # (... to be done)


            # draw time
            time = font.render(str((pygame.time.get_ticks()-start_ticks)//60000%60).zfill(2) + ":" + \
                               str((pygame.time.get_ticks()-start_ticks)//1000%60).zfill(2),True ,WHITE, BLACK)
                                #計時器  前面代表分鐘 get_ticks() 是以毫秒為單位 所以分鐘那裡要 // (1000*60)
                                #       後面代表秒 所以 // 1000 但由於時間是 60進制 所以要 % 60
            self.window.blit(time, (0, 580))  #顯示時間
            # ...(to be done)

            pygame.display.update()
        pygame.quit()

if __name__ == "__main__":
    covid_game = Game()
    covid_game.game_run()


```
